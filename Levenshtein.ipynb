```python
import matplotlib.pyplot as plt
# Longitudes de las cadenas (3, 5, 7, 10)
lengths = [3, 5, 7, 10]

# Tiempos de ejecución (en ms)
time_aleatorias = [0.107, 1.5845, 7.6865, 338.039]
time_inversas = [0.004, 0.0586, 0.2293, 57.9366]
time_iguales = [0.0014, 0.0014, 0.0016, 0.0027]

# ---------------------------------------------------------
# Gráfica 1: Cadenas Aleatorias
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(lengths, time_aleatorias, marker='o', color='red', linestyle='-', linewidth=2)

for i, txt in enumerate(time_aleatorias):
    plt.annotate(f"{txt} ms", (lengths[i], time_aleatorias[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('1. Cadenas Aleatorias: Tiempo de Ejecución vs Longitud')
plt.xlabel('Longitud de la Cadena (caracteres)')
plt.ylabel('Tiempo (ms)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('cadenas_aleatorias.png', bbox_inches='tight') # Guarda la imagen
plt.show()

# ---------------------------------------------------------
# Gráfica 2: Cadenas Inversas
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(lengths, time_inversas, marker='s', color='blue', linestyle='-', linewidth=2)

for i, txt in enumerate(time_inversas):
    plt.annotate(f"{txt} ms", (lengths[i], time_inversas[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('2. Cadenas Inversas: Tiempo de Ejecución vs Longitud')
plt.xlabel('Longitud de la Cadena (caracteres)')
plt.ylabel('Tiempo (ms)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('cadenas_inversas.png', bbox_inches='tight')
plt.show()

# ---------------------------------------------------------
# Gráfica 3: Cadenas Iguales
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(lengths, time_iguales, marker='^', color='green', linestyle='-', linewidth=2)

for i, txt in enumerate(time_iguales):
    plt.annotate(f"{txt} ms", (lengths[i], time_iguales[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('3. Cadenas Iguales: Tiempo de Ejecución vs Longitud')
plt.xlabel('Longitud de la Cadena (caracteres)')
plt.ylabel('Tiempo (ms)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('cadenas_iguales.png', bbox_inches='tight')
plt.show()
```
